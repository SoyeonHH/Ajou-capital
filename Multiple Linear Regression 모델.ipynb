#  SK하이닉스

### 데이터셋 분리

```python
x = SK하이닉스[['시가','종가','평균거래량','웹','유튜브']]
y = SK하이닉스[['다음주 종가']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
x_valid, x_test, y_valid, y_test = train_test_split(x_test, y_test, train_size=0.5, test_size=0.5)
```

### Linear Resgression 모델 학습

```python
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
y_predicted = linear.predict(x_test)
```

```python
plt.plot(y_test.values, label="Actual Next Price")
plt.plot(y_predicted, label="Predicted Next Price")
plt.legend()
plt.show()
```

```python
# 회귀계수
print(linear.coef_)
```

### Model Acurracy Evaluation

```python
r2 = linear.score(x_train,y_train)
r2
```
