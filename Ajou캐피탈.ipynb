# Linear regression

```python
com = SK하이닉스

x = com[['시가','종가','평균거래량','평균가', '기관','외인','웹','유튜브','네이버_검색량']]
y = com[['다음주 종가']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.9, test_size=0.1)

linear = LinearRegression()
linear.fit(x_train,y_train)
y_predicted = linear.predict(x_test)

plt.plot(y_test.values, label="Actual Next Price", c='black')
plt.plot(y_predicted, label="Predicted Next Price", c='purple')
plt.title("Multiple Linear Regression - Prediction")
plt.legend()
plt.show()
```
